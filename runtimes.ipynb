```python
import pandas as pd 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16, 8]
```

```python

headers = ["recursion_limit", "run_time"]

df1 = pd.read_csv('quick_sort40000.csv', names=headers)
df2 = pd.read_csv('quick_sort80000.csv', names=headers)
df3 = pd.read_csv('quick_sort160000.csv', names=headers)
df4 = pd.read_csv('quick_sort320000.csv', names=headers)

df4.head(50)
```

```python

min_val_loc_df1 = tuple(df1.loc[df1['run_time'].idxmin()])
min_val_loc_df2 = tuple(df2.loc[df2['run_time'].idxmin()])
min_val_loc_df3 = tuple(df3.loc[df3['run_time'].idxmin()])
min_val_loc_df4 = tuple(df4.loc[df4['run_time'].idxmin()])
print(min_val_loc_df1)
print(min_val_loc_df2)
print(min_val_loc_df3)
print(min_val_loc_df4)
```

```python

ax1 = df1.plot(title='Recursion Runs', x='recursion_limit', y='run_time', label='size=4000',colormap='Paired')
ax1.annotate(str(min_val_loc_df1),min_val_loc_df1,arrowprops=dict(facecolor='black', shrink=0.01))

df2.plot(ax=ax1,x='recursion_limit', y='run_time',label='size=8000')
ax1.annotate(str(min_val_loc_df2),min_val_loc_df2,arrowprops=dict(facecolor='black', shrink=0.01))

df3.plot(ax=ax1,x='recursion_limit', y='run_time',label='size=160000')
ax1.annotate(str(min_val_loc_df3),min_val_loc_df3,arrowprops=dict(facecolor='black', shrink=0.01))


df4.plot(ax=ax1,x='recursion_limit', y='run_time',label='size=320000')
ax1.annotate(str(min_val_loc_df4),min_val_loc_df4,arrowprops=dict(facecolor='black', shrink=0.01))
plt.xlabel('recursion_limits')
plt.ylabel('run_times')
# plt.show()
```
